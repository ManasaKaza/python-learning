```python
#importing json module to work with json files
import json

#importing difflib module to search for the word when it is given with any little error
from difflib import get_close_matches

data = json.load(open("data.json"))

print("Welcome to your mini dictionary :)\n")

#function to search for a word
def translate(word):
    word = word.lower()
    if word in data:
        return data[word]
    elif word.title() in data:
        return data[word.title()]
    elif word.upper() in data:
        return data[word.upper()]
    elif len(get_close_matches(word , data.keys())) > 0 :
        print("\ndid you mean %s instead" %get_close_matches(word, data.keys())[0])
        decide = input("\npress y for yes or n for no")
        if decide == "y":
            return data[get_close_matches(word , data.keys())[0]]
        elif decide == "n":
            return("Sorry, the word you want isn't available :(\n")
        else:
            return("Invalid key, enter y or n\n")
    else:
        print("Sorry, the word you want isn't available :(\n")

#function to search for the other word if he wants to
def second_choice(ch):
    if ch == "y":
        word = input("Enter the word you want to search\n\n")
        output = translate(word)
        if type(output) == list:
            for item in output:
                print(item)
        else:
            print(output)

    elif ch == "n":
        print("Bye!!")
        return

    elif ch != "y" or "n":
        print("Invalid key, enter y or n")
        word = input("Enter the word you want to search\n\n")
        output = translate(word)
        if type(output) == list:
            for item in output:
                print(item)
        else:
            print(output)
        
    #asking the user if he wants to search again    
    choice = input("\nDo you want to search any other word: y-Yes or n-No\n").lower()
    if choice == "y":
        second_choice(choice)
        
    elif choice == "n":
        print("Bye!!")
        return
    
    elif choice != "y" or "n":
        print("Invalid key, enter y or n")
        word = input("Enter the word you want to search\n\n")
        output = translate(word)
        if type(output) == list:
            for item in output:
                print(item)
        else:
            print(output)
    
#taking user's input        
word = input("Enter the word you want to search\n\n")
output = translate(word)
if type(output) == list:
    for item in output:
        print(item)
else:
    print(output)

#asking the user if he wants to search again     
choice = input("\nDo you want to search any other word: y-Yes or n-No\n").lower()
if choice == "y":
    second_choice(choice)
elif choice == "n":
        print("Bye!!")    
elif choice != "y" or "n":
        opinion = input("Invalid key, enter y or n")
        if opinion == "y":
            word = input("Enter the word you want to search\n\n")
            output = translate(word)
            if type(output) == list:
                for item in output:
                    print(item)
            else:
                print(output)
        else:
            print("Bye!!")
            

    

```
