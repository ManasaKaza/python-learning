```python
#importing module to generate random values
import random

def hangman():
    with open("words.txt", "r") as file:
        text = file.read()
    words = text.split()
    word = random.choice(words)
    length = len(word)
    
    validLetters = 'abcdefghijklmnopqrstuvwxyz'
    turns = length + 5
    guessmade = ''
    
    print("You have to guess the word in ",turns," attempts\n")

    while len(word) > 0:
        main = ""
        missed = 0

        for letter in word:
            if letter in guessmade:
                main = main + letter
            else:
                main = main + "_" + " "
        if main == word:
            print(main)
            print("Hurreyy!! You won")
            break

        print("Guess the word: \n" , main)
        guess = input()

        if guess in validLetters:
            guessmade = guessmade + guess
        else:
            print("Enter a valid character")
            guess = input()

        if guess not in word:
            turns = turns - 1
            if turns > 9:
                print(turns," turns left\n")
            if turns == 9:
                print("9 turns left\n")
                print("  --------  ")
            if turns == 8:
                print("8 turns left\n")
                print("  --------  ")
                print("     O      ")
            if turns == 7:
                print("7 turns left\n")
                print("  --------  ")
                print("     O      ")
                print("     |      ")
            if turns == 6:
                print("6 turns left\n")
                print("  --------  ")
                print("     O      ")
                print("     |      ")
                print("    /       ")
            if turns == 5:
                print("5 turns left\n")
                print("  --------  ")
                print("     O      ")
                print("     |      ")
                print("    / \     ")
            if turns == 4:
                print("4 turns left\n")
                print("  --------  ")
                print("   \ O      ")
                print("     |      ")
                print("    / \     ")
            if turns == 3:
                print("3 turns left\n")
                print("  --------  ")
                print("   \ O /    ")
                print("     |      ")
                print("    / \     ")
            if turns == 2:
                print("2 turns left\n")
                print("  --------  ")
                print("   \ O /|   ")
                print("     |      ")
                print("    / \     ")
            if turns == 1:
                print("1 turns left")
                print("Last breathe counting, Take care!\n")
                print("  --------  ")
                print("   \ O_|/   ")
                print("     |      ")
                print("    / \     ")
            if turns == 0:
                print("Oops!! You lost")
                print("You let a kind man die\n")
                print("  --------  ")
                print("     O_|    ")
                print("    /|\      ")
                print("    / \     \n")
                break
            
    print("The word is: ", word)

    
#function to play other game if the user wants to
def second_choice(ch):
    if ch == "y":
        hangman()
        
    elif ch == "n":
        print("Bye!!")
        return

    elif ch != "y" or "n":
        opinion = input("Invalid key, enter y or n").lower()
        if opinion == "y":
            hangman()
        elif opinion == "n":
            print("Bye!!")
        
    #asking the user if he wants to play again    
    choice = input("\nDo you want to play again: y-Yes or n-No\n").lower()
    if choice == "y":
        second_choice(choice)
        
    elif choice == "n":
        print("Bye!!")
        return
    
    elif choice != "y" or "n":
        opinion = input("Invalid key, enter y or n").lower()
        if opinion == "y":
            hangman()
        elif opinion == "n":
            print("Bye!!")
    
name = input("Enter your name ")
print("Helloo \n" , name )
print("-------------------\n")
hangman()
print()

#asking the user if he wants to play again
choice = input("Do you want to play again: y-Yes or n-No").lower()
if choice == "y":
    hangman()
elif choice == "n":
    print("Bye!!")
else:
    opinion = input("Invalid key, enter y or n").lower()
    if opinion == "y":
        hangman()
    elif opinion == "n":
        print("Bye!!")
```
