```python
df = pd.DataFrame(np.random.rand(100,5), columns=['a','b','c','d','e'])
df
```

```python
df1 = pd.DataFrame({'x':[3,8,1,5,7], 'y':[1,2,3,4,5], 'z':[6,7,8,4,3]})
df1
```

# point plots

```python
df.iplot(kind='scatter')
```

```python
df.iplot(kind='scatter',x='a',y='b')
```

```python
df.iplot(kind='scatter',x='a',y='b',mode='markers')
```

```python
df.iplot(kind='scatter',x='a',y='b',mode='markers',size=10)
```

```python
df.iplot(kind='scatter',x='a',y='b',mode='markers',size=10,symbol='x', colorscale='paired')
```

```python
df.scatter_matrix()
```

# surface plots

```python
df1.iplot(kind='surface')
```

```python
cf.datagen.sinwave(10,.25).iplot(kind='surface',theme='solar',colorscale='brbg',margin=(0,0,0,0))
```
