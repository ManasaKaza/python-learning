# sub plots using object oriented

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(x,y,x,x)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(2,2)
axes[0].plot(x,y)
axes[1].plot(x,y,x,x)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(2,2)
axes[(1,1)].plot(x,y)
axes[(0,1)].plot(x,y,x,x)
plt.tight_layout()
```

# FIgure size

```python
fig = plt.figure(figsize = (5,5))
axes = fig.add_axes([0,0,1,1])
plt.plot(x,y)
```

```python
fig, axes = plt.subplots(2,1,figsize=(10,5))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig = plt.figure(figsize = (5,5))
axes = fig.add_axes([0,0,1,1])
axes.xlabel('xaxis')
plt.plot(x,y)
```

```python
fig = plt.figure(figsize = (5,5))
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('xaxis')
axes.set_ylabel('yaxis')
axes.set_title('title')
plt.plot(x,y)
```

```python
fig = plt.figure()
x1 = fig.add_axes([0.1,0.1,0.8,0.8])
plt.plot(x,y)
x2 = fig.add_axes([0.4,0.4,0.3,0.3])
x1.set_xlabel('x1')
x1.set_ylabel('y1')
x2.set_xlabel('x2')
x2.set_ylabel('y2')
x1.set_title('g1')
x2.set_title('g2')
plt.plot(x,y)
```

```python
fig, axes = plt.subplots(2,1,figsize=(10,5))
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_xlabel('x1')
axes[0].set_ylabel('y1')
axes[1].set_xlabel('x2')
axes[1].set_ylabel('y2')
axes[0].set_title('g1')
axes[1].set_title('g2')
```

# styling plots

```python
fig = plt.figure(figsize = (5,5))
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('xaxis')
axes.set_ylabel('yaxis')
axes.set_title('title')
plt.plot(x,y,color='red',linewidth=5,alpha=0.1,linestyle='dashdot')
```

# Markers

```python
fig = plt.figure(figsize = (5,5))
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('xaxis')
axes.set_ylabel('yaxis')
axes.set_title('title')
plt.plot(x,y, marker='o',markersize=10,markerfacecolor='red',markeredgecolor='green',markeredgewidth=10)
```

# limits

```python
fig = plt.figure(figsize = (5,5))
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('xaxis')
axes.set_ylabel('yaxis')
axes.set_title('title')
axes.set_xlim([0,5])
axes.set_ylim([0,5])
plt.plot(x,y)
```
