# Merging

```python
df1 = pd.DataFrame({'key1':[1,2,3], 'a':[2,3,4], 'b':[5,6,7]})
df2 = pd.DataFrame({'key1':[1,2,3], 'c':[1,2,9], 'd':[5,8,9]})
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2, how='right')
```

```python
pd.merge(df1,df2, how='left')
```

```python
pd.merge(df1,df2, how='inner')
```

```python
pd.merge(df1,df2, how='outer')
```

```python
pd.merge(df1,df2, how='inner', on='key1')
```

```python
pd.merge(df1,df2, how='outer', on='key1')
```

```python
pd.merge(df1,df2, how='left', on='key1')
```

```python
pd.merge(df1,df2, how='right', on='key1')
```

```python
df1 = pd.DataFrame({'key1':[1,2,3], 'a':[2,3,4], 'b':[5,6,7], 'key2':[5,3,2]})
df2 = pd.DataFrame({'key1':[1,2,3], 'c':[1,2,9], 'd':[5,8,9], 'key2':[5,3,2]})
```

```python
pd.merge(df1,df2, how='right', on='key1')
```

```python
pd.merge(df1,df2, how='right', on=['key1','key2'])
```
